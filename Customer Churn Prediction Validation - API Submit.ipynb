```python
df = pd.read_csv(r"Customer Data 2019 to 2020 Validation.csv")
```

```python
cat_df=df.select_dtypes(include = ['object'])
```

```python
num_issue = ['rev_rentals',
'tot_rev_in',
'tot_rev_out',
'tot_rev',
'mon_bal',
'tue_bal',
'wed_bal',
'fri_bal',
'sat_bal',
'sun_bal',
'mon_rev',
'tue_rev',
'wed_rev',
'thu_rev',
'fri_rev',
'sat_rev',
'sun_rev',
'rev_subscription_post',
'tot_rev_in_zur',
'tot_ict_out',
'tot_ict_in',
'rev_mtd',
'rev_prev_month_1',
'rev_prev_month_2',
'rev_prev_month_3']
```

```python
for col in num_issue:
    df[col] = df[col].str.replace(' ','')
    df[col] = df[col].str.replace('\s+','')
    df[col] = df[col].str.replace(',','.')
```

```python
for col in num_issue:
    df[col] = df[col].astype('float')
```

```python
null_cat = ['customer_type','b2b_type','cons_type','tariff_type','hs_make','hs_model','rev_mtd_current']
df[null_cat]= df[null_cat].fillna('None')
df.drop('channel_id', axis = 1, inplace = True)
```

```python
num_df=df.select_dtypes(exclude = ['object'])
```

```python
null_num = ['rev_other','thu_bal','fri_rchg','rev_ibf','rev_mtd','rev_prev_month_1','rev_prev_month_2','rev_prev_month_3']
```

```python
df[null_num]=df[null_num].fillna(df.median())
```

```python
num_df=df.select_dtypes(exclude = ['object'])
```

```python
cat_useless = ['country','opco_name','opco_business_type','hs_make', 'hs_model' , 'datausage30days', 'rev_mtd_current']
df.drop(cat_useless,axis = 1 ,inplace = True)
```

```python
for col in df.select_dtypes(include='object'):
    df[col] = df[col].astype('category')
```
