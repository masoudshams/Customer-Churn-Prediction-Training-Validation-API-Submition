```python
from sklearn.linear_model import LogisticRegression as lr
model = lr()
```

```python
x = df7.drop('status', axis = 1)
y = df7['status']
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x_train, x_test, y_train, y_test = tts(x, y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.preprocessing import MinMaxScaler,RobustScaler,StandardScaler,Normalizer,PowerTransformer,MaxAbsScaler,QuantileTransformer 
```

```python
robust_scaler = RobustScaler()
x_train_r = robust_scaler.fit_transform(x_train)
x_test_r = robust_scaler.transform(x_test) 
```

```python
model.fit(x_train_r, y_train)
```

```python
pred = model.predict(x_test_r)
```

```python
from sklearn import metrics
```

```python
print(metrics.confusion_matrix(y_test, pred))
```

```python
print("Accuracy:", str(round(metrics.accuracy_score(y_test, pred) * 100, 2)), "%")
print("Precision:", str(round(metrics.precision_score(y_test, pred) * 100, 2)), "%")
print("Recall:", str(round(metrics.recall_score(y_test, pred) * 100, 2)), "%")
print("F1_score", str(round(metrics.f1_score(y_test, pred) * 100, 2)), "%")
```
